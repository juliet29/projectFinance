```python
import pandas as pd
import numpy as np
from collections import OrderedDict
import xlsxwriter
```

```python
# eng, procurement, const base cost 
epc_cost = 455e6

# other payments over course of construction period 
split_costs = {
"int_and_fees" : 56.853e6,
"insurance" : 25e6,
"man_and_over ": 8.15e6
}

# payments at financial close  
fc_costs = {
    "land": 12.5e6,
    "dev_exp": 28.5e6,
    "dev_fees": 14.7e6,
    "close" : 14.392e6,
    "conting": 30.754e6
}


```

```python
# EPC payment schedule 
const_months = list(range(1,25))
epc_sched = OrderedDict()

for i in const_months:
    if i <= 3:
        epc_sched[i] = 0.02
    elif i <= 8: 
        epc_sched[i] = 0.04
    elif i <= 14: 
        epc_sched[i] = 0.06
    elif i <= 18: 
        epc_sched[i] = 0.05
    else: 
        epc_sched[i] = 0.03

# epc_sched
```

```python
epc_sched
```

```python
# calculate monthly payments 

epc_pay = [epc_cost * v for v in epc_sched.values()]

split_pay = {}
for k,v in split_costs.items():
    split_pay[k] = [v/len(const_months) for i in const_months]

fc_pay = {}
for k,v in fc_costs.items():
    fc_pay[k] = np.zeros(len(const_months))
    fc_pay[k][0] = v

# fc_pay

assert len(split_pay["insurance"]) == len(epc_pay) == len(fc_pay["land"])

```

```python
# construction model
cm = pd.DataFrame(data=epc_pay, columns=["epc_pay"])
cm1 = pd.concat([cm, pd.DataFrame(split_pay), pd.DataFrame(fc_pay)],  axis=1)
cm1["total_monthly_pay"] = cm1.sum(axis=1)
cme = cm1.T
cme
```
